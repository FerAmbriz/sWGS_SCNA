```python
df_onco = pd.read_csv('/home/fer/Documents/cfDNA/OncoV6/Base_v6.csv')
df_onco.columns = df_onco.iloc[11]
df_onco = df_onco.drop(range(0,12))
df_onco = df_onco.drop(['Type', 'Function', 'ACMG'], axis=1)
#df_onco = df_onco.set_index('ID')
df_cfDNA = pd.melt(df_onco, id_vars=['ID', 'Tier_oncoKB'], value_vars=list(df_onco.columns[1:len(df)]))
df_cfDNA = df_cfDNA.dropna()
df_cfDNA
```

```python
df_cfDNA[df_cfDNA['ID']=='NF1']
```

```python
drugs = pd.read_excel('/home/fer/Documents/cfDNA/ClassifyTier.xlsx')

def match_drugs (gene, oncoKB):
    lst_onco = ['Tier I', 'Tier II']
    if oncoKB in lst_onco:
        drugs_i = drugs[(drugs['Gene']==gene)]
        z = list(drugs_i['drugs_oncoKB'])
        return z[0]
    else:
        return 'NaN'

match_drugs('NF1', 'Tier II')
```

```python
df_cfDNA['drugs'] = list(map(match_drugs, df_cfDNA['ID'], df_cfDNA['Tier_oncoKB']))
df_cfDNA
```

```python
df_cfDNA[df_cfDNA['ID']=='NF1']
```

```python
def match_gene(ID, Gene):
    df_i = crud_data[(crud_data['Sample']== ID) & (crud_data['gene'] == Gene)]
    a = list(df_i['chromosome'])[0]
    b = list(df_i['start'])[0]
    c = list(df_i['end'])[0]
    d = list(df_i['Sample'])[0]
    gene = list(df_i['gene'])[0]
    
    #############
    original = df[(df['ID']== ID) & (df['Gene'] == Gene)]
    e = list(original['TumorType'])[0]
    f = list(original['Cytoband'])[0]
    g = list(original['Type'])[0]
    h = list(original['CN'])[0]   
    
    #############
    df_drugs = df_cfDNA[(df_cfDNA[11] == ID) & (df_cfDNA['ID']== Gene)]
    i = list(df_drugs['Tier_oncoKB'])[0]
    j = list(df_drugs['drugs'])[0]
    return [d,e,a,b,c,f,gene,g,h,i,j]
```

```python
new_df = pd.DataFrame(list(map(match_gene,df['ID'], df['Gene'])))
new_df.columns = ['ID', 'TumorType', 'Chr', 'Start', 'End', 'Cytoband', 'Gene', 'Type', 'CN', 
                  'Tier_oncoKB', 'drugs_oncoKB']
new_df
```

```python
new_df[new_df['ID']=='UEB-0209']
```

```python
df_tum = new_df
```

```python
new_df.to_csv('/home/fer/Documents/cfDNA/Supplementary/CNV_tumor.csv', index = False)
```
