```python
import geopandas as gpd
import pandas as pd
```

```python
# Cargar el conjunto de datos naturalearth_lowres
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
df = pd.read_csv('/home/fer/Downloads/clusters.tsv', sep = '\t')
df
```

```python
lst_lat = [] ; lst_long = []
for country in df['node']:
    country_data = world[world['name'] == country]
    
    if not country_data.empty:
        centroid = country_data.geometry.centroid
        #print(f"Las coordenadas de {country} son {centroid.y.values[0]}, {centroid.x.values[0]}")
        lst_lat.append(centroid.y.values[0]);  lst_long.append(centroid.x.values[0])
    else:
        lst_lat.append('NaN');  lst_long.append('NaN')
        #print(f"No se pudo obtener las coordenadas de {country}")
```

```python
df['latitud'] = lst_lat; df['longitud'] = lst_long
df
```

```python
countries = list(df[df['latitud'] == 'NaN']['node'])
countries
```
