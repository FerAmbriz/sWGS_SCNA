```python
import pandas as pd
```

```python
df = pd.read_csv('/home/fer/Downloads/FinalDataBase_TimeTreat.csv')
#df = df[df['ID_Final']=='UEB-0218']
df
```

```python
df = df.replace({'Se hizo primero la SX':0, 'Se hizo primero la QX': 0, 
            'Falta información':'NaN', 'Se extrajo primero la muestra':1,
            'Se hizo primero la RTX':0
           })
df.head()
```

```python
lst = []
lst2 = []
for i in range(len(df)):
    #cfDNA
    if 1 in list(df[['Time_cfDNA_Cirugia','Time_cfDNA_QX','Time_cfDNA_RXT']].iloc[i]):
        lst.append('Treated Tumor')
    elif 0 in list(df[['Time_cfDNA_Cirugia','Time_cfDNA_QX','Time_cfDNA_RXT']].iloc[i]): 
        lst.append('Primary tratment-naive tumor')
    else:
        lst.append('Unknown')
    # Tumor
    if 1 in list(df[['Time_Tejido_Cirugia','Time_Tejido_QX','Time_Tejido_RXT']].iloc[i]):
        lst2.append('Treated Tumor')
    elif 0 in list(df[['Time_Tejido_Cirugia','Time_Tejido_QX','Time_Tejido_RXT']].iloc[i]):
        lst2.append('Primary tratment-naive tumor')
    else:
        lst2.append('Unknown')
df['StatusTreat_cfDNA'] = lst
df['StatusTreat_Tumor'] = lst2
```

```python
df = df.drop(['StatusTreat'], axis=1)
```
