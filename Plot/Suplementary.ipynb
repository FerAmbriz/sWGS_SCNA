## Match oncoKB

```python
def match_oncoKB_TumorType(ID,gene,typ):
    if 0 == len(list(data_onco[(data_onco['Gene']== gene)]['oncoKB/Tier'])):
        oncoKB = 'III'
    else:
        oncoKB = list (data_onco[(data_onco['Gene']== gene)]['oncoKB/Tier'])[0]
    if typ == 'cfDNA':
        tumorType = list(data_clinic[data_clinic['ID_Final']==ID]['StatusTreat_cfDNA'])[0]
    else:
        tumorType = list(data_clinic[data_clinic['ID_Final']==ID]['StatusTreat_Tumor'])[0]
    return [oncoKB, tumorType]
match_oncoKB_TumorType('UEB-0205', 'CBFB', 'tumor')
```

```python
typ = ['cfDNA']* len(df_cfdna) 
x = list(map(match_oncoKB_TumorType,df_cfdna['ID'],df_cfdna['Gene'],typ))
x = pd.DataFrame(x)

df_cfdna['oncoKB/Tier'] = x[0]
df_cfdna['TumorType'] = x[1]

typ = ['Tumor']* len(df_tumor) 
x = list(map(match_oncoKB_TumorType,df_tumor['ID'],df_tumor['Gene'],typ))
x = pd.DataFrame(x)

df_tumor['oncoKB/Tier'] = x[0]
df_tumor['TumorType'] = x[1]
```
