```python
df_bd = pd.DataFrame()

for i, j, t, m, h in zip(df.columns, stage, treat, sub_molec, histo):
    df_i = pd.DataFrame(df[i].value_counts())
    df_i.columns = ['Count']
    df_i['ID'] = [i] * len(df_i)
    df_i['Stage'] = [j] * len(df_i)
    df_i['Treatment'] = [t] * len(df_i)
    df_i['MolecularSubtype'] = [m] * len(df_i)
    df_i['HistologicalSubtype'] = [h] * len(df_i)
    
    df_bd = pd.concat([df_bd, df_i])
    
df_bd.head()
```

```python
df = pd.read_csv('/home/fer/Documents/cfDNA/Tumor/MutationalSignatures.csv')
df = df.drop(range(0, 4))
df = df.set_index('Unnamed: 0')
df
```

```python
def round_custom(num):
    if num > 0:
        return 1
    else:
        return 0
round_custom(0.1)
```

```python
for i in df.columns:
    df[i] = list(map(float, df[i]))
    df[i] = list(map(round_custom, df[i]))
df
```

```python
firmas_paciente = list(df.sum())
def Average(lst):
    return sum(lst) / len(lst)
print(f'median: {Average(firmas_paciente)}, std: {np.std(firmas_paciente)}')
```
