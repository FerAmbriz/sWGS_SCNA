```python
ID = ID[5:len(ID)]
```

```python
lst_loss_tumor = []
lst_loss_cfDNA = []
lst_gain_tumor = []
lst_match = []
merge = 0
for i in range(5, len(df.columns)+5, 1):
    loss_tumor = 0
    loss_cfDNA = 0
    gain_tumor = 0
    match = 0
    for j in df[i]:
        if j == 'loss_tumor;loss_cfDNA':
            loss_tumor = loss_tumor + 1
            loss_cfDNA = loss_cfDNA + 1
            match = match + 1
        elif j == 'loss_tumor;' or j == 'loss_tumor; ' or j == ' loss_tumor;' or j == 'loss_tumor':
            loss_tumor = loss_tumor + 1
        elif j == 'gain_tumor;' or j == 'gain_tumor; ' or j == ' gain_tumor;' or j == 'gain_tumor':
            gain_tumor = gain_tumor + 1
        elif j == ';loss_cfDNA' or j == ' ;loss_cfDNA' or j == ';loss_cfDNA ' or j == 'loss_cfDNA':
            loss_cfDNA = loss_cfDNA + 1
            
    lst_loss_tumor.append(loss_tumor)
    lst_loss_cfDNA.append(loss_cfDNA)
    lst_gain_tumor.append(gain_tumor)
    lst_match.append(match)
```

```python
df = pd.read_csv('/home/fer/Documents/cfDNA/OncoprintAll_mergeDirectionality/Base_v4.csv', header = None)
df = df[df[0]=='SNV']

df.drop([0, 1, 2, 3], axis = 1, inplace = True)
#df.drop(list(range(0, 12)), axis = 0, inplace = True)
df = df.set_index(4)
df = df.fillna(0)
df
```

```python
lst_nonsynonymous = []
lst_hotspot = []
lst_frameshift = []

for i in range(5, len(df.columns)+5, 1):
    nonsynonymous = 0
    hotspot = 0
    frameshift = 0
    for j in df[i]:
        if j == 'nonsynonymous':
            nonsynonymous = nonsynonymous + 1
        elif j == 'hotspot':
            hotspot = hotspot + 1
        elif j == 'frameshift':
            frameshift = frameshift + 1
    lst_nonsynonymous.append(nonsynonymous)
    lst_hotspot.append(hotspot)
    lst_frameshift.append(frameshift)
```

```python
df_count = pd.DataFrame({
    'ID': ID,
    'loss_tumor': lst_loss_tumor,
    'gain_tumor': lst_gain_tumor,
    'loss_cfDNA': lst_loss_cfDNA,
    'nonsynonymous': lst_nonsynonymous,
    'hotspot': lst_hotspot,
    'frameshift': lst_frameshift,
    'match': lst_match
})
df_count
```

```python
df_count['All'] = df_count['loss_tumor'] + df_count['gain_tumor'] + df_count['loss_cfDNA'] + df_count['nonsynonymous'] + df_count['hotspot'] +  df_count['frameshift'] 
df_count = df_count.sort_values(['All'])
df_count = df_count.drop(['All'], axis=1)
df_count = df_count.set_index('ID')
```
