```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#df = pd.read_csv('/home/fer/Documents/cfDNA/Tumor/TMB.csv')
df = pd.read_csv('/home/fer/Documents/cfDNA/Tumor/TMB.csv')

df= df.sort_values(['TMB/Kb'], ascending=False)
#df['TMB/Kb'] = df['Freq'] / 800
#df.columns = ['ID', 'Freq', 'KB', 'TMB/Kb']
#df.head()

#df = df.set_index('ID')
#df.to_csv('/home/fer/Documents/cfDNA/Tumor/TMB.csv')
df
```

```python
plt.figure(figsize=(13, 7))

ax = sns.barplot(data=df, x='ID_FINAL', y='TMB/Kb', palette='viridis')
ax.axhline(df['TMB/Kb'].mean(),
          color = "red",lw = 2, linestyle ="--")
plt.text(32, 0.8 , 'xÌ„ = ' + str(round(df['TMB/Kb'].mean(), 5)), fontsize = 12)

ax.tick_params(axis='x', rotation=90)
plt.savefig("/home/fer/TMB_viridis.pdf", dpi = 300, pad_inches=0.2, bbox_inches='tight')
```

```python
plt.figure(figsize=(13, 7))

ax = sns.barplot(data=df, x='ID', y='TMB/Kb', palette='magma')
ax.axhline(df['TMB/Kb'].mean(),
          color = "red",lw = 3, linestyle ="--")
plt.text(32, 0.8 , 'xÌ„ = ' + str(round(df['TMB/Kb'].mean(), 5)), fontsize = 12)

ax.tick_params(axis='x', rotation=90)
plt.savefig("/home/fer/TMB_magma.pdf", dpi = 300, pad_inches=0.2, bbox_inches='tight')
```
