```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('/home/fer/Documents/cfDNA/OncoprintAll_mergeDirectionality/Base_v5 _GradientStages.csv')
df = df.drop(['Unnamed: 0', 'Unnamed: 1', 'Unnamed: 2'], axis = 1)
df.columns = df.loc[11]

sub_molec = pd.DataFrame(df.loc[0])


df = df.drop(list(range(0,12)))

df
```

```python
sub_molec = sub_molec.reset_index()
sub_molec.columns = ['ID', 'Type']
sub_molec
```

```python
# DataPreparation
df['Cytoband'] = list(map(str, df['Cytoband']))

def cut_stringchr (string):
    x1= string.split('p')
    x2 = string.split('q')
    if len(x1) < len(x2):
        x = x2
    else:
        x = x1
    return x[0]

def insert_chr (string):
    x = 'chr' + str(string)
    return x
```

```python
df['Cytoband'] = list(map(cut_stringchr, df['Cytoband']))
df['Cytoband'] = list(map(insert_chr, df['Cytoband']))
df
```

```python
ID_2 = list(df.columns.values)
ID_2 = ID_2[2:len(ID_2)]
df_melt = pd.melt(df, id_vars=['Cytoband', 'ID'], value_vars = ID_2)
df_melt.columns = ['chr', 'gene', 'ID', 'CNV']
df_melt
```

```python
df_melt = df_melt[df_melt['CNV'].notnull()]
df_melt
```

```python
df_melt['gene'].value_counts()
```

```python
lst = []
for i,j in zip(sub_molec.ID, sub_molec.Type):
    for k in df_melt.ID:
        if i == k:
            lst.append(j)
len(lst)
```

```python
df_melt['Type'] = lst
df_melt.head()
```

```python
df_melt = df_melt.drop(['ID'], axis = 1)
df_melt
```

```python
df_c = pd.DataFrame(df_melt.groupby(['chr', 'gene']).value_counts())
#df_c[df_c[0] > 1]
df_c.columns = ['Count']
df_c
```
