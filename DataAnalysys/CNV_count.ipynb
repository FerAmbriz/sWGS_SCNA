# Construcción de un DF con caracteristicas clinicas

```python
df = pd.read_csv('/home/fer/Documents/cfDNA/OncoprintAll_merge/Base_v4.csv', header = None)
ID = list(df.iloc[11,:])
Age = list(df.iloc[10,:])
Stadium = list(df.iloc[9,:])
Sex = list(df.iloc[8,:])
ResponseQX = list(df.iloc[7,:])
Histological = list(df.iloc[6,:])
Tumor_progression = list(df.iloc[5,:])
M = list(df.iloc[4,:])
N = list(df.iloc[3,:])
cfDNA_concentration = list(df.iloc[2,:])
TumorFraction = list(df.iloc[1,:])
SubMolec = list(df.iloc[0,:])

df = df[df[0]=='CNV']

df.drop([0, 1, 2, 3], axis = 1, inplace = True)
#df.drop(list(range(0, 12)), axis = 0, inplace = True)
df = df.set_index(4)
df = df.fillna(0)
df
```

```python
ID = ID[5:len(ID)]

Age = Age[5:len(Age)]
Stadium = Stadium[5:len(Stadium)]
Sex = Sex[5:len(Sex)]
ResponseQX = ResponseQX[5:len(ResponseQX)]
Histological = Histological[5:len(Histological)]
Tumor_progression = Tumor_progression[5:len(Tumor_progression)]
M = M [5:len(M)]
N = N[5:len(N)]
TumorFraction = TumorFraction[5:len(TumorFraction)]
SubMolec = SubMolec[5:len(SubMolec)]
cfDNA_concentration = cfDNA_concentration[5:len(cfDNA_concentration)]
```
