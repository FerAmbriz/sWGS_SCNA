```python
import pandas as pd
import requests
import json
import io
from tqdm import tqdm
```

```python
url = 'https://genome.ucsc.edu/cgi-bin/hgTables?hgsid=1442153227_FWCo6wJtrFjEzVt07A5mEs5LeL3m'    
session = requests.Session()

params = {
        'hgsid': '1442153227_FWCo6wJtrFjEzVt07A5mEs5LeL3m',
        'jsh_pageVertPos': '0',
        'clade': 'mammal',
        'org': 'Human',
        'db': 'hg19',
        'hgta_group': 'genes',
        'hgta_track': 'refSeqComposite',
        'hgta_table': 'ncbiRefSeq',
        'hgta_regionType': 'genome',
        'hgta_outputType': 'primaryTable',
        'boolshad.sendToGalaxy': '0',
        'boolshad.sendToGreat': '0',
        'boolshad.sendToGenomeSpace': '0',
        'hgta_outFileName': '',
        'hgta_compressType': 'none',
        'hgta_doTopSubmit': 'get output'
    }

response = session.post(url, data=params)
df = pd.read_csv(io.StringIO(response.text), sep='\t')
df
```

```python
df = df.drop_duplicates(subset=["name2"])
df
```

```python
bed_genes = pd.read_csv('/home/fer/Documents/cfDNA/Tumor/GenesPanel.csv')
bed_genes
```

```python
df_bd = pd.DataFrame()
for i in bed_genes.Gene:
    if i not in list(df['name2']):
         print('No estuvo', i)
    else:
        df_i = df[df['name2']==i]
        df_bd = pd.concat([df_bd, df_i])

df_bd
```

```python
bed_tumor = pd.DataFrame({'chr':df_bd['chrom'], 'start': df_bd['txStart'], 'end': df_bd['txEnd']})
bed_tumor
```

```python
bed_tumor = bed_tumor.set_index('chr')
bed_tumor
```
