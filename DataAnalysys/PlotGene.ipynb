```python
import pandas as pd
import seaborn as sns
```

# Final

```python
df = pd.read_csv('/home/fer/Documents/cfDNA/OncoprintAll_mergeDirectionality/Base_v5.csv')
df = df.drop(['Unnamed: 0', 'Unnamed: 1', 'Unnamed: 2'], axis = 1)
df = df.drop(list(range(0, 12, 1)))
df
```

```python
df['Unnamed: 3'] = list(map(str, df['Unnamed: 3']))
def cut_stringchr (string):
    x1= string.split('p')
    x2 = string.split('q')
    if len(x1) < len(x2):
        x = x2
    else:
        x = x1
    return x[0]
```

```python
df['Unnamed: 3'] = list(map(cut_stringchr, df['Unnamed: 3']))
df
```

```python
def insert_chr (string):
    x = 'chr' + str(string)
    return x
df['Unnamed: 3'] = list(map(insert_chr, df['Unnamed: 3']))

df
```

```python
df.columns = list(range(0, len(df.columns.values)))
df
```

```python
df = df.set_index([0,1])
df = df.T
df
```

```python
df_bd = pd.DataFrame()

for i in df.columns.values:
    df_i = pd.DataFrame(df[i].value_counts()).T 
    df_bd = pd.concat([df_bd, df_i])
df_bd
```

```python
pd.DataFrame(df_bd.T.sum())
```

```python
df_bd_tumor = df_bd.drop([';loss_cfDNA'], axis = 1)
df_bd_tumor
```

```python
df_total_tumor = pd.DataFrame(df_bd_tumor.T.sum())
df_total_tumor = df_total_tumor.reset_index()
df_total_tumor.columns = ['chr', 'gene', 'CNV count']
df_total_tumor['Type'] = ['Tumor'] * len(df_total_tumor)
df_total_tumor
```

```python
df_bd_cfDNA = df_bd.drop(['gain_tumor;', 'loss_tumor;'], axis = 1)
df_bd_cfDNA
```

```python
df_total_cfDNA = pd.DataFrame(df_bd_cfDNA.T.sum())
df_total_cfDNA = df_total_cfDNA.reset_index()
df_total_cfDNA.columns = ['chr', 'gene', 'CNV count']
df_total_cfDNA['Type'] = ['cfDNA'] * len(df_total_cfDNA)
df_total_cfDNA
```

```python
df_final = pd.concat([df_total_tumor, df_total_cfDNA])
df_final
```

```python
import matplotlib.patches as mpatches

def plot_chr (df, chrom):
    df_f = df[df['chr'] == chrom]
    df_f_cfDNA = df_f[df_f['Type'] == 'cfDNA']
    total = df_f.groupby('gene')['CNV count'].sum().reset_index()
    
    ax = sns.barplot(x = 'gene', y = 'CNV count', estimator=sum, data = total, color = "#ff9f68")
    ax = sns.barplot(x = 'gene', y = 'CNV count',  estimator=sum, data = df_f_cfDNA, color = "#48cf90")
    ax.set(ylabel = "Frequency CNV")
    ax.legend(handles=[mpatches.Patch([], [], color="#ff9f68", label='CNV Tumor'), 
                   mpatches.Patch([], [], color="#48cf90", label='CNV cfDNA')
                  ])
    ax.set(title=chrom)
    ax.tick_params(axis='x', rotation=90)
    fig = ax.get_figure()
    fig.savefig('/home/fer/Variants_'+chrom+'.pdf', dpi = 3000, pad_inches=0.2, bbox_inches='tight') 
    return ax
```

```python
plot_chr(df_final, 'chr19')
#plt.savefig('/home/fer/Variants_chr19.pdf', dpi = 3000)
```

```python
plot_chr(df_final, 'chr16')
```

```python
plot_chr(df_final, 'chr1')
```
