```python
import pandas as pd
```

```python
df = pd.read_csv('/home/fer/Documents/cfDNA/OncoprintMerge/Base.v2.csv', header = None)
df
```

```python
df = df.drop(list(range(0,11, 1)),axis=0)
df
```

```python
df.columns = list(df.loc[11])
df = df.drop(11,axis=0)
df
```

```python
df = df.drop(['Type', 'Function', 'Cytoband', 'Status'], axis=1)
df
```

```python
# Tumor
columnas = df.columns
genes = list(df['ID'])

i = 1
df_bd = pd.DataFrame()

while i < len(columnas):
    lst_i = []
    for j in df[columnas[i]]:
        if j == 'loss_tumor;loss_cfDNA' or j == 'gain_tumor;' or j == 'loss_tumor;' or j == 'loss' or j =='gain' or j == 'nonsynonymous SNV' or j == 'HotSpot':
            lst_i.append(1)
        else:
            lst_i.append(0)
    df_bd[i] = lst_i
    i=i+1

df_bd.insert(0, "ID", genes)
df_bd.columns = columnas
df_tumor = df_bd
df_tumor
```

```python
# cfDNA
i = 1
df_bd = pd.DataFrame()

while i < len(columnas):
    lst_i = []
    for j in df[columnas[i]]:
        if j == 'loss_tumor;loss_cfDNA' or j == ';loss_cfDNA' or j == ';gain_cfDNA':
            lst_i.append(1)
        else:
            lst_i.append(0)
    df_bd[i] = lst_i
    i=i+1

df_bd.insert(0, "ID", genes)
df_bd.columns = columnas
df_cfDNA = df_bd
df_cfDNA
```

```python
df_tumor.to_csv('/home/fer/Documents/cfDNA/OncoprintMerge/TumorFilterv2.csv')
df_cfDNA.to_csv('/home/fer/Documents/cfDNA/OncoprintMerge/cfDNAFilterv2.csv')
```
